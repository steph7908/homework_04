## Age Demographics

* Establish bins for ages


* Categorize the existing players using the age bins. Hint: use pd.cut()


* Calculate the numbers and percentages by age group


* Create a summary data frame to hold the results


* Optional: round the percentage column to two decimal points


* Display Age Demographics Table


```python
#left value exclusive, right value inclusive
age_bins = [-1, 9, 14, 19, 24, 29, 34, 39, 100]
age_labels = ['0-9','10-14','15-19','20-24','25-29','30-34','35-39', '40+']
```

```python
purchase_data['Age Bin'] = pd.cut(purchase_data['Age'], age_bins, labels=age_labels)
# purchase_data.head()

grouped_age_bin_df = purchase_data.groupby('Age Bin')
age_count = grouped_age_bin_df['SN'].nunique()
perc_players = age_count / age_count.sum() * 100

combine_df2 = pd.concat([age_count,perc_players], axis=1)
combine_df2.columns = ['Total Count', 'Percentage of Players']
combine_df2
```

## Purchasing Analysis (Age)

* Bin the purchase_data data frame by age


* Run basic calculations to obtain purchase count, avg. purchase price, avg. purchase total per person etc. in the table below


* Create a summary data frame to hold the results


* Optional: give the displayed data cleaner formatting


* Display the summary data frame

```python
#left value exclusive, right value inclusive
age_bins = [-1, 9, 14, 19, 24, 29, 34, 39, 100]
age_labels = ['0-9','10-14','15-19','20-24','25-29','30-34','35-39', '40+']
```

```python
grouped_age_bin_df = purchase_data.groupby('Age Bin')

purch_count = grouped_age_bin_df['Purchase ID'].count()
avg_purch_price = grouped_age_bin_df['Price'].mean()
tot_purch_value = grouped_age_bin_df['Price'].sum()

grouped_age_bin_df['SN'].nunique()
avg_by_age = tot_purch_value / grouped_age_bin_df['SN'].nunique()

combine_df3 = pd.concat([purch_count,avg_purch_price, tot_purch_value, avg_by_age], axis=1)
combine_df3.columns = ['Purchase Count', 'Average Purchase Price', 'Total Purchase Value', 'Avg Total Purchase per Person']
combine_df3
```

## Top Spenders

```python
grouped_id_data = purchase_data.groupby('SN')

purch_sn = grouped_id_data['Purchase ID'].count()
tot_purch_sn = grouped_id_data['Price'].sum()
avg_purch_price = tot_purch_sn / purch_sn
#avg_purch_price['Lisosia93']

combine_df4 = pd.concat([purch_sn,avg_purch_price, tot_purch_sn], axis=1)
combine_df4.columns = ['Purchase Count', 'Average Purchase Price', 'Total Purchase Value']
# combine_df4
combine_df4.sort_values('Total Purchase Value', ascending=False).head()
```
